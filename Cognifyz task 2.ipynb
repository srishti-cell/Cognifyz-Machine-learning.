```python
# importing libraries and dataset
import pandas as pd
df = pd.read_csv("Dataset.csv")
df["Cuisines"]=df["Cuisines"].fillna("Unknown")
rows,columns = df.shape
df.head()
```

```python
# there are some coloumn values which has some invalid characters "?"
accented_columns = ['Restaurant Name','City','Address','Locality','Locality Verbose','Cuisines']
from unidecode import unidecode
def clean_text(text):
    if isinstance(text,str):
        return unidecode(text)
    return text
df[accented_columns]=df[accented_columns].applymap(clean_text)
df.tail()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Read the CSV file into a DataFrame
df1 = pd.read_csv("cleaned1.csv")

# Extracting features
y = df1['Restaurant Name']
x_text = df1['Cuisines'] + " " + df1['Rating color'] + " " + df1['Rating text'] + " " + df1["Votes"].astype(str) + " " + df1["Price range"].astype(str)

# Create a TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Transform the text data using TF-IDF
x = tfidf.fit_transform(x_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarityScore = cosine_similarity(x)
similarityScore
```

```python
userInput = input("Enter Your Preferred Cuisines :")
cuisineList = df1["Cuisines"]
for i in range(len(cuisineList)):
    if userInput in cuisineList[i]:
        k = df1.iloc[i]['Restaurant Name']
        index = i
        break
RecommendationScore = list(enumerate(similarityScore[index]))        
```

```python
# Recommendation of Resturant
sortedResturantList = sorted(RecommendationScore,key=lambda x:x[1],reverse = True)
count = 0
i=0
for rest in sortedResturantList:
    Index = rest[0]
    k=df1.iloc[Index]["Restaurant Name"]
    print(k)
    count+=1
    
    if count==10:
        break
    i+=1    
```
